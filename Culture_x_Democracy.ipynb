**Correlação entre Cultura e Democracia: Regressão Linear**

```
import pandas as pd

from google.colab import files
uploaded = files.upload()
```

**Base de Dados**

```
import io
database = pd.read_csv(io.BytesIO(uploaded['cult-democracy.csv']))
print(database)
```

**Análise de Correlação entre as dimensões culturais PDI x IDV**


```
import numpy as np
import matplotlib.pyplot as plt

plt.plot(database['Country'],database['pdi'],label='PDI')
plt.plot(database['Country'],database['idv'],label='IDV')
plt.ylabel('Degree')
plt.title('PDI x IDV')
plt.legend()
plt.xticks(np.arange(min(database['PDI']), max(database['Overall'])+80, 3))
plt.xticks(rotation=90, fontsize=10)
plt.show()
```

**Correlação entre a dimensão cultural PDI x Média dos 05 indicadores de Democracia**





```
plt.plot(database['Country'],database['PDI'], label='PDI')
plt.plot(database['Country'],database['Overall'], label='Overall')
plt.xticks(np.arange(min(database['PDI']), max(database['Overall'])+80, 3))
plt.xticks(rotation=90, fontsize=10)
plt.ylabel('Degree')
plt.title('PDI x Democracy Overall')
plt.legend(loc='upper right')
plt.show()
```

**Correlação entre a dimensão cultural IDV x Média dos 05 indicadores de Democracia**



```
plt.plot(database['Country'],database['IDV'],label='IDV')
plt.plot(database['Country'],database['Overall'],label='Overall')
plt.xticks(np.arange(min(database['IDV']), max(database['Overall'])+80, 3))
plt.xticks(rotation=90, fontsize=10)
plt.ylabel('Degree')
plt.title('IDV x Democracy Overall')
plt.legend(loc='upper right')
plt.show()
```
