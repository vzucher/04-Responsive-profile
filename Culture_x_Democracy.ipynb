**Análise de Correlação entre Cultura e Democracia: Regressão Linear**

```
import pandas as pd

from google.colab import files
uploaded = files.upload()
```

```
import io
database = pd.read_csv(io.BytesIO(uploaded['cult-democracy.csv']))
print(database)
```

```
import numpy as np
import matplotlib.pyplot as plt

plt.plot(database['Country'],database[['pdi', 'idv']])
plt.xticks(np.arange(min(database['PDI']), max(database['Overall'])+80, 3))
plt.xticks(rotation=90, fontsize=10)
plt.show()
```
